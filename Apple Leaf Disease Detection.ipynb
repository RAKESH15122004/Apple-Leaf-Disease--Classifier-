```python
import os
import shutil
import uuid
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, BatchNormalization, Dense, GlobalAveragePooling2D, Input

# Setup paths
base_dir = f"temp_dataset_{uuid.uuid4().hex}"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
os.makedirs(train_dir)
os.makedirs(test_dir)

dataset_paths = [
    "/content/Alternaria leaf spot",
    "/content/Brown spot",
    "/content/Gray spot",
    "/content/Healthy leaf",
    "/content/Rust"
]

# Data preparation
def collect_and_split(dataset_paths, train_dir, test_dir, split_ratio=0.8):
    image_paths, labels = [], []
    for class_dir in dataset_paths:
        class_name = os.path.basename(class_dir)
        if not os.path.isdir(class_dir):
            continue
        for img_file in os.listdir(class_dir):
            if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(class_dir, img_file))
                labels.append(class_name)

    train_paths, test_paths, train_labels, test_labels = train_test_split(
        image_paths, labels, train_size=split_ratio, stratify=labels, random_state=42)

    def copy_files(paths, labels, target_base):
        for img_path, label in zip(paths, labels):
            dest_dir = os.path.join(target_base, label)
            os.makedirs(dest_dir, exist_ok=True)
            shutil.copy(img_path, os.path.join(dest_dir, os.path.basename(img_path)))

    copy_files(train_paths, train_labels, train_dir)
    copy_files(test_paths, test_labels, test_dir)

collect_and_split(dataset_paths, train_dir, test_dir)

# Image generators
img_size = 128
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical'
)
test_data = test_datagen.flow_from_directory(
    test_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical', shuffle=False
)

# Model architecture
model = tf.keras.models.Sequential([
    Input(shape=(img_size, img_size, 3)),

    Conv2D(32, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.2),

    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.3),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.4),

    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.4),

    GlobalAveragePooling2D(),

    Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.5),
    Dense(train_data.num_classes, activation='softmax')
])

# Model summary
model.summary()

# Compile model
model.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# Train model
history = model.fit(train_data, epochs=10, validation_data=test_data)

# Save model
model.save("leaf_disease_model.h5")
print("💾 Model saved as 'leaf_disease_model.h5'")

# Evaluate model
loss, acc = model.evaluate(test_data)
print(f"\n✅ Final Test Accuracy: {acc:.4f}")

# Predictions
y_pred_probs = model.predict(test_data)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_data.classes
class_labels = list(test_data.class_indices.keys())

# Classification report
print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_labels))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Accuracy & loss curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import tensorflow as tf

# Load model (adjust path if needed)
model = tf.keras.models.load_model("leaf_disease_model.h5")

# Define class labels
class_names = ["Apple Scab", "Black Rot", "Cedar Apple Rust", "Healthy"]  # Example

def predict(image):
    try:
        # Preprocess image
        image = image.resize((224, 224))  # Match your model input
        img_array = np.array(image) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # Predict
        preds = model.predict(img_array)
        predicted_index = np.argmax(preds)
        confidence = float(np.max(preds))

        # Get class label
        diagnosis = class_names[predicted_index]

        # Recommend an action (example logic)
        if diagnosis == "Healthy":
            action = "No action needed."
        else:
            action = "Apply appropriate treatment and monitor plant health."

        return diagnosis, f"{confidence:.2%}", action

    except Exception as e:
        # Print the error for debugging in Colab
        print("ERROR in prediction:", str(e))
        return "Error", "Error", "Error"
```

```python
#5.  EfficientNetB0

import os
import shutil
import uuid
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, precision_score,
    recall_score, f1_score, fbeta_score
)
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Input
from tensorflow.keras.optimizers import Adam

# Setup paths
base_dir = f"temp_dataset_{uuid.uuid4().hex}"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

dataset_paths = [
    "/content/Alternaria leaf spot",
    "/content/Brown spot",
    "/content/Gray spot",
    "/content/Healthy leaf",
    "/content/Rust"
]

# Data preparation
def collect_and_split(dataset_paths, train_dir, test_dir, split_ratio=0.8):
    image_paths, labels = [], []
    for class_dir in dataset_paths:
        class_name = os.path.basename(class_dir)
        if not os.path.isdir(class_dir):
            continue
        for img_file in os.listdir(class_dir):
            if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(class_dir, img_file))
                labels.append(class_name)

    train_paths, test_paths, train_labels, test_labels = train_test_split(
        image_paths, labels, train_size=split_ratio, stratify=labels, random_state=42)

    def copy_files(paths, labels, target_base):
        for img_path, label in zip(paths, labels):
            dest_dir = os.path.join(target_base, label)
            os.makedirs(dest_dir, exist_ok=True)
            shutil.copy(img_path, os.path.join(dest_dir, os.path.basename(img_path)))

    copy_files(train_paths, train_labels, train_dir)
    copy_files(test_paths, test_labels, test_dir)
    return train_paths, test_paths, train_labels, test_labels

train_paths, test_paths, train_labels, test_labels = collect_and_split(dataset_paths, train_dir, test_dir)

# Image generators
img_size = 128
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
test_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical'
)
test_data = test_datagen.flow_from_directory(
    test_dir, target_size=(img_size, img_size), batch_size=batch_size, class_mode='categorical', shuffle=False
)

# Model architecture: EfficientNetB0
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(img_size, img_size, 3))
base_model.trainable = False  # Freeze base

model = tf.keras.models.Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(train_data.num_classes, activation='softmax')
])

model.summary()

# Compile model
learning_rate = 0.0001
optimizer = Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer,
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# Train model
epochs = 10
start_time = time.time()
history = model.fit(train_data, epochs=epochs, validation_data=test_data)
end_time = time.time()
execution_time_ms = (end_time - start_time) * 1000

# Save model
model.save("leaf_disease_model_efficientnetb0.h5")
print("💾 Model saved as 'leaf_disease_model_efficientnetb0.h5'")

# Evaluate model
loss, acc = model.evaluate(test_data)
print(f"\n✅ Final Test Accuracy: {acc:.4f}")

# Predictions
y_pred_probs = model.predict(test_data)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_data.classes
class_labels = list(test_data.class_indices.keys())

# Classification report
report = classification_report(y_true, y_pred, target_names=class_labels, output_dict=True)
print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_labels))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels, cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Accuracy & loss curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.show()

# --- Metrics Calculation ---
try:
    aucs = roc_auc_score(test_data.labels, y_pred_probs, multi_class='ovr', average=None)
    auc_macro = roc_auc_score(test_data.labels, y_pred_probs, multi_class='ovr', average='macro')
    auc_weighted = roc_auc_score(test_data.labels, y_pred_probs, multi_class='ovr', average='weighted')
except Exception as e:
    aucs = auc_macro = auc_weighted = "N/A (not enough samples per class)"

precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
f2 = fbeta_score(y_true, y_pred, beta=2, average='macro', zero_division=0)
fbeta = fbeta_score(y_true, y_pred, beta=0.5, average='macro', zero_division=0)

TP = np.diag(cm)
FP = cm.sum(axis=0) - TP
FN = cm.sum(axis=1) - TP
TN = cm.sum() - (FP + FN + TP)

sensitivity = TP / (TP + FN + 1e-8)
specificity = TN / (TN + FP + 1e-8)
miss_rate = FN / (FN + TP + 1e-8)
ppv = TP / (TP + FP + 1e-8)
npv = TN / (TN + FN + 1e-8)
fpr = FP / (FP + TN + 1e-8)
fnr = FN / (FN + TP + 1e-8)
lpr = sensitivity / (1 - specificity + 1e-8)
lnr = (1 - sensitivity) / (specificity + 1e-8)
fm_index = np.sqrt(precision * recall)

print("\n--- Model & Dataset Parameters ---")
print(f"1. Sample Size: {len(train_paths) + len(test_paths)}")
print(f"2. Train Dataset: {len(train_paths)} | Test Dataset: {len(test_paths)}")
print(f"3. Proposed Model: EfficientNetB0-based CNN")
print(f"4. Sample Size: {len(train_paths) + len(test_paths)}")
print(f"5. Input Size: {img_size}x{img_size}x3")
print(f"6. Hidden Layer Size: 256")
print(f"7. Number of Neurons in Hidden Layer: 256")
print(f"8. Activation Function of Hidden Layer: relu")
print(f"9. No of Neurons in Output Layer: {train_data.num_classes}")
print(f"10. Activation Function of Output Layer: softmax")
print(f"11. Optimizer: Adam")
print(f"12. Loss Function: categorical_crossentropy")
print(f"13. Performance Metrics: accuracy, precision, recall, F1, AUC")
print(f"14. No of Epochs: {epochs}")
print(f"15. Learning Rate: {learning_rate}")
print(f"16. Execution time (in ms): {execution_time_ms:.2f}")
print(f"17. AUCs (per class): {aucs}")
print(f"    Macro AUC: {auc_macro}")
print(f"    Weighted AUC: {auc_weighted}")
print(f"18. Training Accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"19. Testing Accuracy: {acc:.4f}")
print(f"20. Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")
print(f"21. Loss: {loss:.4f}")
print(f"22. Validation Loss: {history.history['val_loss'][-1]:.4f}")
print(f"23. Precision (macro): {precision:.4f}")
print(f"24. Recall (macro): {recall:.4f}")
print(f"25. Classification Miss Rate (macro): {miss_rate.mean():.4f}")
print(f"26. Sensitivity (macro): {sensitivity.mean():.4f}")
print(f"27. Specificity (macro): {specificity.mean():.4f}")
print(f"28. F1-Score (macro): {f1:.4f}")
print(f"29. F2-Score (macro): {f2:.4f}")
print(f"30. FBeta-Score (beta=0.5, macro): {fbeta:.4f}")
print(f"31. Positive Predicted Value (macro): {ppv.mean():.4f}")
print(f"32. Negative Predicted Value (macro): {npv.mean():.4f}")
print(f"33. False Positive Ratio (macro): {fpr.mean():.4f}")
print(f"34. False Negative Ratio (macro): {fnr.mean():.4f}")
print(f"35. Likelihood Positive Ratio (macro): {lpr.mean():.4f}")
print(f"36. Likelihood Negative Ratio (macro): {lnr.mean():.4f}")
print(f"37. Fowlkes-Mallows index (macro): {fm_index:.4f}")
```
